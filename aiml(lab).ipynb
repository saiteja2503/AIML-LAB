```python
from queue import Queue
graph={
    0:[1,2,3],
    1:[0,2],
    2:[1,3,5],
    3:[0,2,4],
    4:[3,5,7],
    5:[2,4,6,7],
    6:[5,7],
    7:[]
}
print("the adjancecy list representing the graph is:")
print(graph)
def bfs(graph,source):
  Q = Queue()
  visited_vertices=set()
  Q.put(source)
  visited_vertices.update({source})
  while not Q.empty():
    vertex=Q.get()
    print(vertex,end="-->")
    for u in graph[vertex]:
      if u not in visited_vertices:
        Q.put(u)
        visited_vertices.update({u})

bfs(graph,0)
```

```python
from queue import Queue
graph1={
    'A': ['B','C','G'],
    'B': ['A'],
    'C': ['A','D','F'],
    'D': ['A','C','E','H'],
    'E': ['D','H','F'],
    'F': ['C','E'],
    'G': ['A','D','H'],
    'H': ['D','E','G']
}
def dfs(graph,node,visited):
  if node not in visited:
    visited.append(node)
    for k in graph[node]:
      dfs(graph,k,visited)
  return visited
visited=dfs(graph1,'B',[])
print(visited)
```
