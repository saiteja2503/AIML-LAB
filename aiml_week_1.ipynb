##Lab2 task is to visualize the datasets or data using matplotlib builtin function graphs

```python
people=["saiteja","jenny","ramu","rajesh","nick"]
age=[1,2,3,4,5]
weight=[22,33,44,55,66]
height=[123,124,125,126,132]
import matplotlib.pyplot as plt
plt.scatter(weight,height)
plt.title("Relationship between weight and height")
plt.xlabel("weight in kgs")
plt.ylabel("height in meters")
plt.show()
plt.scatter(weight,age)
plt.title("Relationship between age and weight")
plt.xlabel("weight in kgs",fontsize=15)
plt.ylabel("age in years",fontsize=15)
plt.show()
plt.bar(people,age)
plt.title("Relationship between people and age")
plt.xlabel("people")
plt.ylabel("age")
plt.show
plt.figure(figsize=(7,5))
plt.hist(weight,color='red',edgecolor='k',alpha=0.75,bins=5)
plt.title("Histogram of  patient weight",fontsize=18)
plt.xlabel("weight in kgs",fontsize=15)
plt.ylabel("frequency")
plt.show()
```
